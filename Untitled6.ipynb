**Reasoning**:
The subtask is to implement the 'Listar Personas' functionality by modifying the existing menu loop to display all created Persona objects with an index when option '2' is selected.



````python
class Persona:
  def __init__(self, nombre, edad):
    self.nombre = nombre # atributo de instancia
    self.edad = edad # otro atributo de instancia
  def saludar(self):
    print(f"Hola, soy {self.nombre} y tengo {self.edad} años")

personas = []

while True:
    print("\n--- Menú Principal ---")
    print("1. Crear nueva Persona")
    print("2. Listar Personas")
    print("3. Hacer que una Persona salude")
    print("4. Salir")

    opcion = input("Por favor, elige una opción: ")

    if opcion == '1':
        print("Has elegido crear una nueva Persona.")
        nombre = input("Introduce el nombre de la persona: ")
        edad_valida = False
        while not edad_valida:
            try:
                edad_str = input(f"Introduce la edad de {nombre}: ")
                edad = int(edad_str)
                if edad > 0:
                    edad_valida = True
                else:
                    print("La edad debe ser un número positivo. Intenta de nuevo.")
            except ValueError:
                print("Entrada no válida. Por favor, introduce un número entero para la edad.")

        nueva_persona = Persona(nombre, edad)
        personas.append(nueva_persona)
        print(f"¡{nombre} ha sido creada exitosamente y añadida a la lista!")
    elif opcion == '2':
        print("\n--- Lista de Personas ---")
        if not personas:
            print("No hay personas creadas aún.")
        else:
            for i, persona in enumerate(personas):
                print(f"{i+1}. Nombre: {persona.nombre}, Edad: {persona.edad}")
    elif opcion == '3':
        print("Has elegido hacer que una Persona salude.")
    elif opcion == '4':
        print("Saliendo del programa. ¡Hasta luego!")
        break
    else:
        print("Opción no válida. Por favor, intenta de nuevo.")
````

````python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def saludar(self):
        print(f"Hola, soy {self.nombre} y tengo {self.edad} años")

    def cumplir_anios(self):
        self.edad += 1
        print(f"{self.nombre} ahora tiene {self.edad} años")


# Lista donde se guardan las personas
personas = []


def mostrar_menu():
    print("\n--- MENÚ DE PERSONAS ---")
    print("1. Crear nueva persona")
    print("2. Listar personas")
    print("3. Saludar a todas las personas")
    print("4. Aumentar edad de una persona")
    print("5. Eliminar una persona")
    print("6. Salir")


def crear_persona():
    nombre = input("Ingrese el nombre: ")
    edad = int(input("Ingrese la edad: "))
    persona = Persona(nombre, edad)
    personas.append(persona)
    print("Persona creada correctamente")


def listar_personas():
    if not personas:
        print("No hay personas registradas")
    else:
        print("\nLista de personas:")
        for i, persona in enumerate(personas):
            print(f"{i + 1}. {persona.nombre} - {persona.edad} años")


def saludar_todos():
    if not personas:
        print("No hay personas para saludar")
    else:
        print("\nSaludos:")
        for persona in personas:
            persona.saludar()


def aumentar_edad():
    listar_personas()
    if personas:
        opcion = int(input("Seleccione el número de la persona: "))
        if 1 <= opcion <= len(personas):
            personas[opcion - 1].cumplir_anios()
        else:
            print("Opción inválida")


def eliminar_persona():
    listar_personas()
    if personas:
        opcion = int(input("Seleccione el número de la persona a eliminar: "))
        if 1 <= opcion <= len(personas):
            eliminada = personas.pop(opcion - 1)
            print(f"Persona '{eliminada.nombre}' eliminada correctamente")
        else:
            print("Opción inválida")


# Programa principal
while True:
    mostrar_menu()
    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        crear_persona()
    elif opcion == "2":
        listar_personas()
    elif opcion == "3":
        saludar_todos()
    elif opcion == "4":
        aumentar_edad()
    elif opcion == "5":
        eliminar_persona()
    elif opcion == "6":
        print("Saliendo del programa...")
        break
    else:
        print("Opción inválida, intente de nuevo")
````

````python
class Persona:
  def __init__(self, nombre, edad):
    self.nombre = nombre # atributo de instancia
    self.edad = edad # otro atributo de instancia
  def saludar(self):
    print(f"Hola, soy {self.nombre} y tengo {self.edad} años")

personas = []

while True:
    print("\n--- Menú Principal ---")
    print("1. Crear nueva Persona")
    print("2. Listar Personas")
    print("3. Hacer que una Persona salude")
    print("4. Salir")

    opcion = input("Por favor, elige una opción: ")

    if opcion == '1':
        print("Has elegido crear una nueva Persona.")
        nombre = input("Introduce el nombre de la persona: ")
        edad_valida = False
        while not edad_valida:
            try:
                edad_str = input(f"Introduce la edad de {nombre}: ")
                edad = int(edad_str)
                if edad > 0:
                    edad_valida = True
                else:
                    print("La edad debe ser un número positivo. Intenta de nuevo.")
            except ValueError:
                print("Entrada no válida. Por favor, introduce un número entero para la edad.")

        nueva_persona = Persona(nombre, edad)
        personas.append(nueva_persona)
        print(f"¡{nombre} ha sido creada exitosamente y añadida a la lista!")
    elif opcion == '2':
        print("\n--- Lista de Personas ---")
        if not personas:
            print("No hay personas creadas aún.")
        else:
            for i, persona in enumerate(personas):
                print(f"{i+1}. Nombre: {persona.nombre}, Edad: {persona.edad}")
    elif opcion == '3':
        print("Has elegido hacer que una Persona salude.")
    elif opcion == '4':
        print("Saliendo del programa. ¡Hasta luego!")
        break
    else:
        print("Opción no válida. Por favor, intenta de nuevo.")
````
